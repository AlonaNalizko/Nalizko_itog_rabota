Всё про http.


HTTP - это протокол передачи гипертекста, используемый для информации между браузером и серверами. Он работает в режиме запрос-ответ. То есть клиент передаёт на сервер запрос к определённому ресурсу, а сервер отправляет клиенту этот ресурс, либо возвращает ошибку (если ресурс передать нельзя).

Запрос протокола HTTP состоит из метода (определяется операция, которую выполняет клиент), URI (ссылка на сам ресурс, может также содержать параметры), версия HTTP, заголовок (доп информация о запросе, например, информация о том, откуда пришёл запрос, из какого браузера, нужна ли авторизация пользователя для получения информации) и тело(сами данные, который мы передаём, например, ввод логина и пароля при входе).

Адрес (URI) и параметры (Query String) - это конкретное указание, куда именно идет запрос. URI -  это путь к ресурсу на сервере, а Query String - конкретная страница сайта/ручка, уточнение, какую именно информацию пользователь хочет получить.

Методы HTTP:
GET - для получения какой-то информации (обычно в этом запросе нет тела, оно тут не требуется и не используется)
POST - отправка данных на сервер
PUT - обновление существующих данных
PATCH - частичное обновление ресурса на сервере
DELETE - удаление данных

Информация из тела запроса должна считываться только конечным пользователем. Балансировщики не анализируют тело запроса из-за высокой нагрузки, но в некоторых случаях могут читать context-type для маршрутизации или использовать данныйе из тела для сложной логики.
